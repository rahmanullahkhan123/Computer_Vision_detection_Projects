# Indoor Door Detection using YOLOv8
This notebook contains the complete, separated, and corrected workflow to train, validate, and test a YOLOv8 model for door detection.

```python
# ==========================================
# 1. Environment Setup
# ==========================================
# Ensure required packages are updated and fix numpy compatibility issues.
!pip install ultralytics "numpy<2.0.0" scipy --upgrade
```

```python
# ==========================================
# 2. Initialize the YOLOv8 Model
# ==========================================
from ultralytics import YOLO

# Explicitly use an object detection model
print("Loading YOLOv8 object detection model...")
model = YOLO('yolov8n.pt')  # Nano version for fast CPU processing
```

```python
# ==========================================
# 3. Train the Model
# ==========================================
import os

# Resolve the absolute path of the current directory to avoid path errors
data_path = 'd:/Deteaction_Projects/door detection/archive/data.yaml'

print(f"Using dataset config: {data_path}")

# Enforcing task='detect' to prevent accidental segmentation fallback that causes coco dataset downloads
results = model.train(
    data=data_path,
    task='detect',              # FORCE object detection to prevent coco-segment download errors
    epochs=5,                  
    imgsz=640,                  
    batch=16,                   
    device='cpu',               # Running on CPU
    project='Door_Detection',   
    name='Run_1',               
    plots=True,                 
    exist_ok=True               # Overwrite if Run_1 already exists
)
print("Training Complete!")
```

```python
# ==========================================
# 4. Validate the Trained Model
# ==========================================
# Run this cell AFTER training is finished to see accuracy metrics on the validation set.
# If you restarted the environment, load the best model first:
# model = YOLO('Door_Detection/Run_1/weights/best.pt')

print("Validating on val dataset...")
metrics = model.val(
    data='d:/Deteaction_Projects/door detection/archive/data.yaml',
    task='detect',
    project='Door_Detection',
    name='Run_1_Val',
    exist_ok=True
)
print("Validation Complete!")
```

```python
# ==========================================
# 5. Run Inference / Testing & Display Image
# ==========================================
import matplotlib.pyplot as plt
import cv2

print("Running test inference on unseen images...")
test_results = model.predict(
    source='d:/Deteaction_Projects/door detection/archive/test/images',
    conf=0.5,
    save=True,
    project='Door_Detection',
    name='Run_1_Predict',
    exist_ok=True
)

print("\nInference completed. Displaying up to 3 sample images inline:\n")

# Display first few images inline in the notebook
for res in test_results[:3]:
    img_bgr = res.plot()
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize=(8, 8))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```
